```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

import numpy as np

def bernoulli_pdf(x, p):
  """
  Calculates the probability distribution function (PDF) of a Bernoulli distribution.

  Args:
    x: The value of the random variable (0 or 1).
    p: The probability of success (value 1).

  Returns:
    The probability of observing the given value x.
  """
  if x == 1:
    return p
  elif x == 0:
    return 1 - p
  else:
    return 0

# Example usage
probability_of_success = 0.6  # Probability of getting a 1 (success)
probability_of_failure = 1- probability_of_success
x_value = 1  # The value we want to find the probability for

pdf_value = bernoulli_pdf(x_value, probability_of_success)

print(f"The PDF for x = {x_value} is: {pdf_value}")
```

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
# uniform distribution

import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(x, low, high):
    """
    Calculates the cumulative distribution function (CDF) of a discrete uniform distribution.

    Args:
        x: The value at which to evaluate the CDF.
        low: The lowest value in the distribution's range.
        high: The highest value in the distribution's range.

    Returns:
        The CDF value at x.
    """
    if x < low:
        return 0.0
    elif x >= high:
        return 1.0
    else:
        return (x - low + 1) / (high - low + 1)

# Parameters for the discrete uniform distribution
low = 1
high = 6

# Generate x-values for the plot
x_values = np.arange(low - 1, high + 2)

# Calculate CDF values for each x-value
cdf_values = [discrete_uniform_cdf(x, low, high) for x in x_values]

# Plot the CDF
plt.step(x_values, cdf_values, where='post')  # Use 'post' for right-continuous step function
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.grid(True)
plt.show()
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculates Z-scores from a dataset and plots them.

    Args:
        data: A NumPy array or list of data values.
    """

    # Calculate Z-scores
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev

    # Plot the Z-scores
    plt.hist(z_scores, bins=50, density=True, alpha=0.7, edgecolor='black')
    plt.title('Distribution of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.show()

    return z_scores  # Return the calculated Z-scores

# Example usage:
data = np.random.normal(loc=50, scale=10, size=1000)  # Example data
z_scores = calculate_and_plot_zscores(data)
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 10000
num_samples = 1000
sample_size = 30

# Generate a non-normal population (exponential distribution)
population = np.random.exponential(scale=1, size=population_size)

# Calculate the population mean and standard deviation
population_mean = np.mean(population)
population_std = np.std(population)

# Collect sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.hist(sample_means, bins=50, density=True, alpha=0.7, edgecolor='black')
plt.title('Central Limit Theorem Demonstration')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

# Plot the theoretical normal distribution
x = np.linspace(min(sample_means), max(sample_means), 100)
theoretical_pdf = (1 / (population_std / np.sqrt(sample_size) * np.sqrt(2 * np.pi))) * \
                  np.exp(-((x - population_mean)**2) / (2 * (population_std / np.sqrt(sample_size))**2))
plt.plot(x, theoretical_pdf, color='red', linewidth=2, label='Theoretical Normal')

plt.legend()
plt.show()

print(f"Population Mean: {population_mean:.2f}")
print(f"Population Standard Deviation: {population_std:.2f}")
print(f"Mean of Sample Means: {np.mean(sample_means):.2f}")
print(f"Standard Deviation of Sample Means: {np.std(sample_means):.2f}")
```

```python
# 18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
# distribution

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_zscore_and_compare(data_point, data_set):
    """
    Calculates the Z-score for a data point and compares it to a standard normal distribution.

    Args:
        data_point: The data point for which to calculate the Z-score.
        data_set: The dataset to which the data point belongs.
    """

    # Calculate the Z-score
    mean = np.mean(data_set)
    std_dev = np.std(data_set)
    z_score = (data_point - mean) / std_dev

    # Print the Z-score
    print(f"Z-score for data point {data_point}: {z_score:.2f}")

    # Plot the standard normal distribution and the data point's Z-score
    x = np.linspace(-4, 4, 100)
    pdf = stats.norm.pdf(x, loc=0, scale=1)  # Standard normal distribution PDF
    plt.plot(x, pdf, color='blue', linewidth=2, label='Standard Normal Distribution')
    plt.axvline(x=z_score, color='red', linestyle='--', label=f'Data Point Z-score ({z_score:.2f})')
    plt.title('Comparison with Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data_set = np.random.normal(loc=50, scale=10, size=1000)  # Example dataset
data_point = 65  # Example data point

calculate_zscore_and_compare(data_point, data_set)
```

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
import scipy.stats as stats

# Sample data
sample_data = np.array([48, 52, 55, 45, 50, 53, 47, 51, 49, 54])

# Population parameters (assumed known for Z-test)
population_mean = 50  # Hypothesized population mean
population_std = 5    # Population standard deviation

# Hypothesis testing
# Null hypothesis: The sample mean is equal to the population mean (mu = 50)
# Alternative hypothesis: The sample mean is not equal to the population mean (mu != 50)

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Set significance level (alpha)
alpha = 0.05

# Make a decision
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("There is sufficient evidence to suggest that the sample mean is different from the population mean.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is not enough evidence to suggest that the sample mean is different from the population mean.")

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats

# Sample data
data = np.array([25, 30, 35, 32, 28, 31, 29, 33, 34, 27])

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
sample_size = len(data)

# Set confidence level
confidence_level = 0.95  # 95% confidence interval

# Calculate critical value (Z-score)
critical_value = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate margin of error
margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

# Calculate confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Level: {confidence_level * 100:.0f}%")
print(f"Confidence Interval: {confidence_interval}")

# Interpretation
print("\nInterpretation:")
print(f"We are {confidence_level * 100:.0f}% confident that the true population mean lies within the interval {confidence_interval}.")
```

```python
# 25. Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets
dataset1 = np.random.normal(loc=50, scale=10, size=100)  # Normal distribution with mean 50, std 10
dataset2 = np.random.normal(loc=60, scale=5, size=100)   # Normal distribution with mean 60, std 5

# Calculate standard deviations
std_dev1 = np.std(dataset1)
std_dev2 = np.std(dataset2)

# Print standard deviations
print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

# Visualize the difference
plt.figure(figsize=(8, 6))

# Histograms of the datasets
plt.hist(dataset1, bins=20, alpha=0.5, label='Dataset 1', edgecolor='black')
plt.hist(dataset2, bins=20, alpha=0.5, label='Dataset 2', edgecolor='black')

# Vertical lines representing standard deviations
plt.axvline(x=np.mean(dataset1) + std_dev1, color='blue', linestyle='--', label='Dataset 1 Std Dev')
plt.axvline(x=np.mean(dataset1) - std_dev1, color='blue', linestyle='--')
plt.axvline(x=np.mean(dataset2) + std_dev2, color='orange', linestyle='--', label='Dataset 2 Std Dev')
plt.axvline(x=np.mean(dataset2) - std_dev2, color='orange', linestyle='--')

plt.title('Comparison of Standard Deviations')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np

# Generate data from a normal distribution
data = np.random.normal(loc=50, scale=10, size=100)  # Mean 50, Standard Deviation 10, 100 data points

# Calculate the range
data_range = np.max(data) - np.min(data)

# Calculate the IQR
q1 = np.percentile(data, 25)  # First quartile (25th percentile)
q3 = np.percentile(data, 75)  # Third quartile (75th percentile)
iqr = q3 - q1

# Print the results
print(f"Data Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
# 27. Implement Z-score normalization on a dataset and visualize its transformation5

import numpy as np
import matplotlib.pyplot as plt

def zscore_normalize(data):
    """
    Performs Z-score normalization on a dataset.

    Args:
        data: A NumPy array or list of data values.

    Returns:
        A NumPy array containing the normalized data.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = (data - mean) / std_dev
    return normalized_data

# Generate some sample data (e.g., from a normal distribution)
data = np.random.normal(loc=50, scale=10, size=100)

# Perform Z-score normalization
normalized_data = zscore_normalize(data)

# Visualize the transformation
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)  # Subplot for original data
plt.hist(data, bins=20, alpha=0.7, edgecolor='black')
plt.title('Original Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)  # Subplot for normalized data
plt.hist(normalized_data, bins=20, alpha=0.7, edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-scores')
plt.ylabel('Frequency')

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# 28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
# distribution

import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(data):
    """
    Calculates the skewness and kurtosis of a dataset.

    Args:
        data: A NumPy array or list of data values.

    Returns:
        A tuple containing the skewness and kurtosis.
    """
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    return skewness, kurtosis

# Generate data from a normal distribution
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate skewness and kurtosis
skewness, kurtosis = calculate_skewness_kurtosis(data)

# Print the results
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```
